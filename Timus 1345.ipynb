<h4 style="color:MediumBlue;">Результат</h4>
Необходимо вставить теги форматирования, чтобы можно было нужным образом раскрасить исходный текст данной на вход программы при помощи HTML. Для этого следует выполнить следующие условия:
<ol>
<li>Все комментарии должны быть заключены в пару тегов “&lt;span class=comment&gt;” и “&lt;/span&gt;”.</li>
<li>Все ключевые слова должны быть заключены в пару тегов “&lt;span class=keyword&gt;” и “&lt;/span&gt;”.</li>
<li>Все строки должны быть заключены в пару тегов “&lt;span class=string&gt;” и “&lt;/span&gt;”.</li>
<li>Все числа должны быть заключены в пару тегов “&lt;span class=number&gt;” и “&lt;/span&gt;”.</li>
<li>Если ключевые слова, строки или числа попали в комментарии, то они считаются частью комментария и не считаются ключевыми словами, стоками или числами.</li>
<li>Если ключевые слова, комментарии или числа попали в строки, то они считаются частью строки и не считаются ключевыми словами, комментариями или числами.</li>
</ol>
<p><b>Строка</b> — это последовательность символов ограниченная апострофами («’»), не содержащая внутри других апострофов. Либо символ решётка («#»), следом за которым идет непустая последовательность цифр. Во втором случае последовательность необходимо брать максимальную. То есть в последовательности символов “#123” подстрока “#1” не является строкой, в то время, как вся последовательность “#123” строкой является.</p>
<p><b>Числа</b> начинаются с цифры, заканчиваются цифрой и внутри могут содержать лишь цифры и, возможно, одну точку («.»). Так же в случае со строкой, числом называется именно максимальная по включению последовательность символов, удовлетворяющая описанному правилу.</p>
<p><b>Идентификатор</b> может начинаться с буквы либо знака подчёркивания и содержать буквы, цифры и знаки подчёркивания. Аналогично строке и числу, идентификатором называется лишь максимальная по включению, строка, удовлетворяющая данным требованиям.</p>
<p>Есть <b>комментарии</b> двух типов: комментарий первого типа начинается с ‘{’ и оканчивается ‘}’, обе скобки - часть комментария; комментарий второго типа начинается с ‘//’ и оканчивается переводом строки, слэши — часть комментария, а перевод строки — нет. Вложенные комментарии являются частью внешних комментариев.</p>
<p>Ключевые слова - это следующие идентификаторы: 'and', 'array', 'begin', 'case', 'class', 'const', 'div', 'do', 'else', 'end', 'for', 'function', 'if', 'implementation', 'interface', 'mod', 'not', 'of', 'or', 'procedure', 'program', 'record', 'repeat', 'shl', 'shr', 'string', 'then', 'to', 'type', 'unit', 'until', 'uses', 'var', 'with', 'while'.</p>
<p>Тут каждое ключевое слово заключено в апострофы, которые не следует считать частями ключевых слов. Ключевые слова показаны в нижнем регистре, но на самом деле они могут встречаться в тексте в любом регистре. Например, ImPlEmentAtioN — это ключевое слово.</p>
<p>Задача доступна по <a href="https://acm.timus.ru/problem.aspx?space=1&num=1345">ссылке</a></p>

```python
keywords = ('and', 'array', 'begin', 'case', 'class', 
            'const','div', 'do', 'else', 'end', 'for',
            'function', 'if', 'implementation', 'interface', 
            'mod', 'not', 'of', 'or', 'procedure',
            'program', 'record', 'repeat', 'shl', 'shr', 
            'string', 'then', 'to', 'type', 'unit', 'until',
            'uses', 'var', 'with', 'while')

span_class_number = '<span class=number>'
span_class_string = '<span class=string>'
span_class_comment = '<span class=comment>'
span_class_keyword = '<span class=keyword>'
close_span = '</span>'


def push_html_tag(s, tag, index):
    """
    Функция вставляет в строку определенный html тег 
    и возвращает строку с тегом
        
        Параметры:
            s (str): строка, в которую требуется вставить тег
            tag (str): html тег
            index (int): индекс того, куда тег нужно вставить
        Возвращаемые параметры:
            s (str): строка с html тегом 
    """
    s = s[:index] + tag + s[index:]
    return s


program_text = []
program_line = ''

# Этот блок предназначен для того, чтобы считать
# входные данные с Timus, так как другие способы не сработали
while True:
    try:
        program_line = input()
    except Exception:
        break
    program_text.append(program_line)

   
i = 0 
j = 0
new_html_string = '' # Копия исходной строки в, которую вставляются теги
tags_length = 0 # Длинна вставленных в строку тегов

# Основная идея заключается в том, чтобы считывать
# текст программы и посимвольно анализировать его.
while i < len(program_text):
    new_html_string = program_text[i]
    while j < len(program_text[i]):
        
        # Если встречается символ "#", то перед ним ставится
        # открывающих тег строки, а закрывающих ставится после 
        # того, как закончится последовательность цифр.
        if program_text[i][j] == '#':
            new_html_string = push_html_tag(new_html_string,
                                            span_class_string,
                                            j + tags_length)
            tags_length += len(span_class_string)
            j += 1
            while program_text[i][j].isdigit():
                j += 1
            new_html_string = push_html_tag(new_html_string, 
                                            close_span,
                                            j + tags_length)
            tags_length += len(close_span)
            
        # Если встречается символ "'", то перед ним ставится
        # открывающих тег строки, а закрывающих ставится после 
        # того, как встретится еще один такой символ, который укажет,
        # что строки языка Pascal закончилась.
        elif program_text[i][j] == "'":
            new_html_string = push_html_tag(new_html_string,
                                            span_class_string,
                                            j + tags_length)
            tags_length += len(span_class_string)
            j += 1
            while program_text[i][j] != "'":
                j += 1
            new_html_string = push_html_tag(new_html_string,
                                            close_span,
                                            j + tags_length + 1)
            tags_length += len(close_span)
            j += 1
            
        # Если встречается символы "//", то перед ними ставится
        # открывающих тег комментария. Закрывающий тег ставится 
        # в конце строки кода на языке Pascal и анализ текущей 
        # строки прекращается, так как после однострочного 
        # комментария ничего попасться не может.
        elif program_text[i][j] == '/' and program_text[i][j + 1] == '/':
            new_html_string = push_html_tag(new_html_string,
                                            span_class_comment,
                                            j + tags_length)
            new_html_string = new_html_string + close_span
            break
            
        # Если встречается символ "_", то это идентификатор, который
        # не обрамляется тегами.
        elif program_text[i][j] == '_':
            while (program_text[i][j].isalpha() or 
                   program_text[i][j].isdigit() or 
                   program_text[i][j] == '_'):
                j += 1

        # Тег комментария ставится, если встречается символ "{".
        # Далее ищется символ "}", после которого должен стоять 
        # закрывающий тег.
        # Они могут быть на разных строках, поэтому при окончании 
        # анализируемой строки нужно считать следующую строку и обнулить 
        # переменные индекса символа и длинны вставленных тегов.
        elif program_text[i][j] == '{':
            new_html_string = push_html_tag(new_html_string,
                                            span_class_comment,
                                            j + tags_length)
            tags_length += len(span_class_comment)
            while program_text[i][j] != '}':
                j += 1
                if j >= len(program_text[i]):
                    program_text[i] = new_html_string
                    i += 1
                    new_html_string = program_text[i]
                    tags_length = 0
                    j = 0
            new_html_string = push_html_tag(new_html_string,
                                            close_span,
                                            j + tags_length + 1)
            tags_length += len(close_span)

        # Если встретилась буква, то изначально подразумевается, что
        # это ключевое слово. 
        # Если последовательность букв заканчивается символом "_" или
        # числом, то это идентификатор, который не нужно обрамлять
        # тегами. 
        # Если последовательность букв закончилась пробелом или другими
        # символами, кроме указанных выше, то она сравнивается со
        # списком ключевых слов и обрамляется тегами в случае
        # соответствия.
        elif program_text[i][j].isalpha():
            start_keyw_j = j
            is_var = False
            while j < len(program_text[i]) and \
                  program_text[i][j].isalpha() or \
                  is_var:
                j += 1
                if j >= len(program_text[i]):
                    break
                if program_text[i][j].isdigit() or \
                   program_text[i][j] == '_':
                    is_var = True
            if program_text[i][start_keyw_j:j].lower() in keywords and \
               not is_var:
                new_html_string = push_html_tag(new_html_string,
                                                span_class_keyword,
                                                start_keyw_j + tags_length)
                tags_length += len(span_class_keyword)
                new_html_string = push_html_tag(new_html_string,
                                                close_span,
                                                j + tags_length)
                tags_length += len(close_span)
        
        # Если встретилась цифра, то перед ней ставится 
        # открывающий тег числа, а закрывающий ставится
        # тогда, когда последовательность цифр заканчивается.
        #
        # Последовательность может разделяться ТОЛЬКО ОДНОЙ точкой
        elif program_text[i][j].isdigit():
            new_html_string = push_html_tag(new_html_string,
                                            span_class_number,
                                            j + tags_length)
            tags_length += len(span_class_number)
            while program_text[i][j].isdigit() or \
                  program_text[i][j] == '.' and \
                  program_text[i][j+1].isdigit():
                j += 1
            new_html_string = push_html_tag(new_html_string,
                                            close_span,
                                            j + tags_length)
            tags_length += len(close_span)
        else:
            j += 1
            
    program_text[i] = new_html_string
    tags_length = 0
    j = 0
    i += 1
    
    
for i in program_text:
    print(i)
```
